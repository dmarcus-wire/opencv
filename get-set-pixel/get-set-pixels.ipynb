## Import Packages

```python
from matplotlib import pyplot as plt
import argparse
import cv2
```

## Function to display images in Jupyter

```python
def plt_imshow(title, image):
    # convert the image from BGR to RGB color space and display it
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## Get and set pixels with OpenCV

```python
args = {
    "image": "hitchhikers.jpg"
}
```

```python
# load the image and grab spatial dimensions
image = cv2.imread(args["image"])
(h, w) = image.shape[:2]
plt_imshow("Original", image)
```

```python
# images are NumPy arrays with origin at (0,0) top-left
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

# access pixel at x=50 and y=20
(b, g, r) = image[0, 0]
print("Pixel at (50, 20) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

# update the pixel at (50, 20) and set to red
image[20, 50] = (0, 0, 255)
(b, g, r) = image[20, 50]
print("Pixel at (50, 20 - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

```python
# compute the center of the image w / 2 and h / 2
(cX, cY) = (w //2, h // 2)

# using NumPy arrays we can slice to grab regions of interest
# grab top-left
tl = image[0:cY, 0:cX]
plt_imshow("Top-Left Corner", tl)

# grab top-right
tr = image[0:cY, cX:w]
plt_imshow("Top-Right Corner", tr)

# grab bottom-left
bl = image[cY:h, 0:cX]
plt_imshow("Bottom-Left Corner", bl)

# grab bottom-right
br = image[cY:h, cX:w]
plt_imshow("Bottom-Right Corner", br)
```

```python
# set the top-left to green
image[0:cY, 0:cX] = (0, 255, 0)

# show updated image
plt_imshow("Updated", image)
```
